```python
import pandas as pd
import numpy as np
```

```python
#Import the CSV file with all the anime data 

ratings = pd.read_csv('Anime Recommendation System.csv')
```

```python
#I use 45M rows, if your PC can run all the 55M rows, go ahead

ratings_min = ratings.iloc[0:45000000,:]
ratings_min.info()
print(ratings_min.shape)
```

```python
#get rid of multiindex


userRatings = ratings_min.pivot_table(index=['user_id'], columns=['Name'], values=['rating'])

userRatings = userRatings.xs('rating', axis=1, level=0)

userRatings.head()
```

```python
#Correlation Matrix Pearson works well for me

corrMatrix = userRatings.corr(method='pearson')
```

```python
corrMatrix.head()
```

```python
#Find the best anime for you
corrMatrix.loc[:,'Koukaku Kidoutai']
```

```python
#Function for the algorithm

def get_similar(anime_name,rating):
    similar_ratings = corrMatrix[anime_name]*(rating-2.5)
    similar_ratings = similar_ratings.sort_values(ascending=False)
    #print(type(similar_ratings))
    return similar_ratings
```

```python
# I defined 3 users preferences 

user = [("Cowboy Bebop",10), ("Paprika",10), ('Koukaku Kidoutai', 9)]
user2 = [('Kono Subarashii Sekai ni Shukufuku wo!', 10), ('Gintama', 10)]
userH = [('Durarara!!', 11)]
```

```python
similar_movies = pd.DataFrame()

for anime,rating in userH:
    similar_movies = similar_movies.append(get_similar(anime,rating),ignore_index = True)

similar_movies.sum().sort_values(ascending=False).head(20)
```
